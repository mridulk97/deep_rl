```python
class Agent_QLearning:
    def __init__(self):
        self.env = gym.make(ENV_NAME)
        self.state = self.env.reset()
        # self.values = collections.defaultdict(float)
        self.eps = 0.1
        self.train = True
        self.values = np.zeros((observation_space, action_space))
    
    def max_val_for_state(self, state):
        pass
    
    def set_eval_mode(self):
        self.train = False
        
    
    def sample_env(self): 
        """
        Inputs: 
            - self: an agent 
        Returns: 
            - a tuple: (old_state, action, reward, new_reward) 
        """
        
        # getting the best action given the current state following epsilon-greedy method
        if self.train:
            # if the value function is all zeros (initially) choose action randomly
            # if self.best_value_and_action(self.state)[0] == 0: # may not satisfy in the beginning due to float type
            if np.max(self.values[state]) == 0:
                action = self.env.action_space.sample()
                print('here')


            # else choose action based on epsilon-greedy method
            elif np.random.rand() > self.eps:
                # _, action = self.best_value_and_action(self.state)
                action = np.argmax(self.values[state])
                print('action')
            else:
                action = self.env.action_space.sample()
                print('else')

        else:
            # _, action = self.best_value_and_action(self.state)
            action = np.argmax(self.values[state])
        

        # taking the action chosen based on epsilon-greedy method
        new_state, reward, terminal, info = self.env.step(action)
        # print(self.state, action, reward, new_state, terminal)

        return self.state, action, reward, new_state, terminal
        
    
        
    def best_value_and_action(self, state): 
        """
        Inputs: 
            - self: an agent 
            - state: current state
        Returns: 
            - best_value: the best value updated 
            - best_action: the best action taken. 
        """
        best_value = 0
        best_action = 0
        for s, a in self.values.keys():
            if s == state:
                if self.values[(s,a)] >= best_value:
                    best_value = self.values[(s,a)]
                    best_action = a
        return best_value, best_action
    
    def value_update(self, s, a, r, next_s): 
        """
        Inputs: 
            - self: an agent 
            - s: state 
            - a: action 
            - r: reward 
            - next_s: next state
        Returns: 
            - self.values[(s, a)]: the updated value of (s, a).
        """
        # self.values[(s, a)] = self.values[(s, a)] + ALPHA * (r + GAMMA * np.max(self.values[(next_s, :)]) - self.values[(s, a)])
        
        # self.values[(s, a)] = self.values[(s, a)] + ALPHA * (r + GAMMA * self.best_value_and_action(next_s)[1] - self.values[(s, a)])
        print('val_update')
        print(self.values[(s, a)])
        print(s, a, r, next_s)
        self.values[(s, a)] = self.values[(s, a)] + ALPHA * (r + GAMMA * np.max(self.values[next_s, :]) - self.values[(s, a)])
        print(self.values[(s, a)])
        return self.values[(s, a)]
        
        
        
    def play_episode(self, env):
        """ 
        Inputs: 
            - self: an agent 
            - env: the environment
        Returns:
            - total_reward: the total reward after playing an episode
        """
        
        # self.reset_state()

        total_reward = 0

        terminal = False
        while not terminal:

            # taking the best action from the current state based on epsilon-greedy method
            current_state, action, reward, new_state, terminal = self.sample_env()
            # print(current_state, action, reward, new_state, terminal)

            if self.train:
                # updating the value function based on the action chosen greedily
                self.value_update(current_state, action, reward, new_state)

            # updating the current state
            self.state = new_state

            total_reward += reward

        return total_reward
```
